```python
import lpips
import torch
import os
import cv2
import numpy as np
from PIL import Image
```

```python
class Nstr:
    def __init__(self, arg):
       self.x=arg
    def __sub__(self,other):
        c=self.x.replace(other.x,"")
        return c
    
# 加载预训练的LPIPS模型
lpips_model = lpips.LPIPS(net="alex").cuda()
# 假设已经有了要计算LPIPS距离的两张图片 image1 和 image2
# 加载图像文件
path_pre = "/userHome/zy/zpc/joint-De-SR/results/test_RCANx4_RSSCN7_bs4/visualization/RSSCN7/"
path_gt = "/userHome/zy/zpc/jointdatasets/RSSCN7/RSSCN7_test_HR/"

lpips_total = 0
num = 0

for name in os.listdir(path_gt):
    # Load images
    img_pre = lpips.im2tensor(lpips.load_image(path_pre+(Nstr(name)-Nstr(".jpg")+"_test_RCANx4_RSSCN7_bs4.png")))  # RGB image from [-1,1]
    img_gt = lpips.im2tensor(lpips.load_image(path_gt+name))

    img_pre = img_pre.cuda()
    img_gt = img_gt.cuda()
    #     img_res = Image.open(folder_path_res+i)
    #     img_gt = Image.open(folder_path_gt+i)
    #     # 将图像转换为PyTorch的Tensor格式

    #     img_res = torch.tensor(np.array(img_res)).permute(2, 0, 1).unsqueeze(0).float() / 255.0
    #     img_gt = torch.tensor(np.array(img_gt)).permute(2, 0, 1).unsqueeze(0).float() / 255.0

    # 使用LPIPS模型计算距离
    distance = lpips_model(img_pre, img_gt)
    lpips_total += distance.item()
    num += 1
    print("lpips:", distance.item())

this_lpips = lpips_total / num
print("LPIPS distance:", this_lpips)
```
